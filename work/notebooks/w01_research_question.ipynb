## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research question: Which pages should be reviewed first for a content refresh based on observable performance signals?

Decision: Decide which pages to prioritize for review.

Action: Review and update the highest-priority pages first.

Cost of a wrong recommendation: Time may be spent reviewing pages that do not need attention while genuinely declining pages could be missed.

Why ML can help: Multiple signals may interact in ways that are difficult to capture with a simple hand-written rule.
