## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

Rule

My rule prioritizes pages that receive high impressions but have a low click-through rate (CTR) for their average search position. These pages already have good visibility, so improving their title and meta description is likely to increase clicks.

Reason code

CTR_FIX – High impressions with lower-than-expected CTR for the page's average position.

Action label

Improve Title & Meta Description

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(df.head())
print(df.columns)
```

```python
# 1. Fill missing CTR values with 0 so we don't get errors
df['ctr'] = df['ctr'].fillna(0)

# 2. Calculate the baseline score
# We prioritize high volume impressions with low CTR
df['score'] = df['impressions_90d'] * (1 - df['ctr'])

# 3. Create descriptive columns for the action
df['reason_code'] = "CTR_FIX"
df['action'] = "Improve Title & Meta Description"

# 4. Rank by score (highest score = top priority)
df_ranked = df.sort_values(by='score', ascending=False)

# 5. Save the output
import os
os.makedirs("outputs", exist_ok=True)
df_ranked.to_csv("outputs/baseline_action_score.csv", index=False)

# 6. Show the top 5 candidates
print(df_ranked[['content_id', 'impressions_90d', 'ctr', 'score']].head(5))
```

Top-20 Review
