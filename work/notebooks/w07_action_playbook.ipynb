## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Step 2: Inspect the dataset
print(df.shape)
print(df.columns)
print(df.head())

# Step 3: Validate the data
print(df.isnull().sum())
print(df.dtypes)
print(df.describe())

required_cols = [
    "content_id",
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position"
]

missing = [col for col in required_cols if col not in df.columns]
print("Missing columns:", missing)

# Step 4: Create the priority score
df["priority_score"] = df["impressions_90d"] * (1 - df["ctr"])

# Step 5: Sort by priority score
df = df.sort_values(by="priority_score", ascending=False)

# Step 6: Add rank
df["rank"] = range(1, len(df) + 1)

# Display the ranked queue
print(df[["rank", "content_id", "priority_score"]].head(10))
```

## 2. Intended use and limits

*Who uses this, for what — and where it stops being valid.*

This action queue is designed to help content and SEO teams decide which pages should be reviewed first. It provides a simple, data-driven way to prioritize pages that may benefit from updates or improvements.

The recommendations are intended to support human decision-making, not replace it. The results depend on the available search data and the scoring method used. They should be treated as directional guidance rather than final decisions, since content quality, business priorities, and recent events are not fully captured by the data.

## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*

Before making any changes, a person should review the page to make sure the recommendation makes sense. They should check whether the data is recent, confirm the page still matches its purpose, and consider any business or content requirements that are not visible in the dataset.

Important decisions such as rewriting content, deleting pages, changing factual information, or publishing updates should never be fully automated. The action queue should be used as decision-support, with a human making the final decision.

## 4. Monitoring / retrain triggers

*What would tell you the recommendations went stale?*

The recommendations should be reviewed whenever search performance changes noticeably or when new data becomes available. If click-through rate, impressions, or average position show consistent changes over time, the scoring method should be checked and updated if needed.

Regular reviews help keep the action queue useful and ensure that recommendations continue to reflect the latest search performance.
