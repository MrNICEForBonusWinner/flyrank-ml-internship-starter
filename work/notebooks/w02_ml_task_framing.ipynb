## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking

My lane is a ranking problem because the goal is to identify which webpages should be reviewed or refreshed first. Instead of predicting a simple yes/no answer, the model prioritizes pages based on their likelihood of benefiting from an update

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is whether a webpage should be refreshed. Since there is no direct label saying "refresh this page," a proxy target can be created using observed performance signals such as traffic changes or engagement metrics. This provides a measurable outcome for the model to learn from.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only consider one or two conditions at a time, such as page age or traffic. In reality, many factors interact, making the decision more complex. Machine learning can learn patterns across multiple features and produce a better ranking than a single hand-written rule.
