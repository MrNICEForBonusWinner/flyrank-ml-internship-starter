## 1. Build the feature vector

*Code that actually builds it — engineered features, categorical handling, fills.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")


df = df.drop(columns=["content_id", "client_id"])


X = df.drop(columns=["impressions_90d"])
y = df["impressions_90d"]


missing_values = X.isnull().sum()

print(missing_values)


print(X.dtypes)


for column in X.columns:

    # Numeric Features
    if X[column].dtype in ["int64", "float64"]:
        X[column] = X[column].fillna(X[column].median())

    # Categorical Features
    elif X[column].dtype == "object":
        X[column] = X[column].fillna(X[column].mode()[0])


unordered_columns = [
    "content_type",
    "main_intent",
    "provider_used",
    "model_used"
]


X = pd.get_dummies(
    X,
    columns=unordered_columns,
    dtype=int
)

print(X.head())
print(X.shape)

X.isnull().sum()
```

## 2. Feature notes (meaning, missing, categorical, available-when?)

*For each feature: what it means, how missing values are handled, and whether it exists BEFORE the moment you predict.*

The feature vector contains content performance, SEO, engagement, and content metadata. Numeric features such as search volume, CPC, CTR, average position, clicks, sessions, and pageviews were retained. Missing numeric values were filled using the median, while missing categorical values were filled using the most frequent category (mode). Unordered categorical features (content_type, main_intent, provider_used, and model_used) were converted using One-Hot Encoding. All selected features are available before prediction, making them suitable as model inputs.

## 3. The leakage hunt

*Attack your own features: label-derived columns, future windows, product flags. Show the test.*

I checked the selected features for possible data leakage by reviewing whether any feature directly contained or revealed the target variable (impressions_90d). Identifier columns (content_id and client_id) were removed because they do not provide useful predictive information and could cause memorization. No columns that directly represented the target value were intentionally included. The remaining features are treated as predictive inputs rather than label-derived information.

## 4. What I excluded and why

*The list of fields you refused to use — with one line of why each.*

I excluded content_id and client_id during feature selection because they are identifier columns rather than predictive features. These columns uniquely identify records or clients but do not describe the content itself. Including them could encourage the model to memorize specific identifiers instead of learning general patterns that apply to unseen data.
